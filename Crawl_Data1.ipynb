```python
from facebook_scraper import get_posts
import numpy as np
import pandas as pd
```

DEFINE VARIABLES

```python
FANPAGE_LINK = "ESPN"
FOLDER_PATH = "Data/"
COOKIE_PATH = "ckies.txt"
PAGE_NUMBER = 10
```

GET POSTS FROM FANPAGE

```python
post_list = []
for post in get_posts(FANPAGE_LINK, options ={"comment": True, "reactions": True, "allow_extra_requests": True},
                      extra_info = True, pages = PAGE_NUMBER, cookies = COOKIE_PATH):
    print(post)
    post_list.append(post)
```

Convert into dataframe

```python
post_df = pd.DataFrame(columns=post_list[0].keys(),index=range(len(post_list)),data = post_list)
#save in a csv file
path = FOLDER_PATH + FANPAGE_LINK +".csv"
post_df.to_csv(path,index=False) 
```
